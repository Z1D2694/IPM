```python
import pandas as pd
```

```python
def resultados(data:pd.DataFrame) -> pd.DataFrame:
    filtro = data[data['Turner_flag']==1]

    filtro = filtro.sort_values(by='AOF [MMCD]', ascending=False)

    df = filtro[['Pws [kg/cm2]','AOF [MMCD]']
                ].groupby('Pws [kg/cm2]').max()

    df = df.reset_index().set_index(['Pws [kg/cm2]','AOF [MMCD]'])

    filtro = filtro[['Pws [kg/cm2]','AOF [MMCD]', 'Qgas [MMCD]', 
                    'Qlig [MCD]', 'Pwf [kg/cm2]','dp_Fric [kg/cm2]', 
                        'dp_grav [kg/cm2]']
                    ].set_index(['Pws [kg/cm2]','AOF [MMCD]'],drop=True)

    resultado = pd.concat([df,filtro], axis=1, join='inner')
    
    return resultado 
```

```python
data = pd.read_excel('Solucion_ID_tbg_2.441_in_prof.xlsx')
```

```python
resultados(data)
```
